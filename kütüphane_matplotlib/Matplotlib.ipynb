 <span style="color:purple">
`ax = fig.add_axes([left, bottom, width, height]):` 

Figüre bir eksen ekler ve eksenin figürdeki konumunu belirler. left, bottom, width ve height değerleri, eksenin figür içinde kapladığı alanı belirler. Bu değerler 0 ile 1 arasında olmalıdır ve figürün yüzde cinsinden konum ve boyutunu ifade eder.


* left: Eksenin figürün sol kenarından uzaklığı.
* bottom: Eksenin figürün alt kenarından uzaklığı.
* width: Eksenin genişliği.
* height: Eksenin yüksekliği.

</span> 

```python
#az önceki plot' a veri ekledik
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

### Daha Kolay ve Daha Sağlam Yöntem: plt.subplots() Kullanımı
</span> 

<span style="color:purple">
    
Bu yöntem, tek bir satırda hem figürü hem de eksenleri oluşturmanızı sağlar ve genellikle tercih edilen yöntemdir.

__Adımlar:__
* __fig, ax = plt.subplots():__ Yeni bir figür ve eksen(ler) oluşturur. Bu yöntem, hem figürü (fig) hem de eksenleri (ax) döner ve bu da daha kompakt ve okunabilir bir kod sağlar.
* __ax.plot(x, y):__ Eksen üzerinde x ve y verileriyle bir çizgi grafiği çizer.
* __plt.show():__ Grafiği görüntüler.
</span> 

```python
fig, ax = plt.subplots(figsize = (4,4))
ax.plot(x, [50, 100, 200, 250]);
```

# 2. En Yaygın Grafik Türlerini NumPy Dizileri Kullanarak Oluşturma

</span>

```python
import numpy as np
```

<span style="color:blue">

## 1. Çizgi Grafiği (Line Plot) 

</span>

<span style="color:green">

Çizgi grafiği, veri noktalarını bir çizgiyle birleştirir ve verilerin zaman içindeki değişimini göstermek için yaygın olarak kullanılır.

</span>



```python
x = np.linspace(0, 10, 100)
x[:10]
```

<span style="color:purple">

- **x = np.linspace(0, 10, 100):** 0 ile 10 arasında 100 eşit aralıklı değer içeren bir NumPy dizisi oluşturur. linspace fonksiyonu, belirli bir aralıktaki sayıları eşit aralıklarla oluşturmak için kullanılır.
- **x[:10]:** x dizisinin ilk 10 elemanını gösterir. Bu adım, diziye genel bir bakış sağlar.

</span>


```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(x, x**2);
```

<span style="color:purple">

- **fig, ax = plt.subplots():** Yeni bir figür ve eksen(ler) oluşturur. Bu, çizim alanını belirler.

- __ax.plot(x, x**2):__ x değerleri ile x'in karesi olan y değerleri arasında bir çizgi grafiği çizer. x**2 ifadesi, x değerlerinin karesini alır.

</span>


<span style="color:blue">

## 2. Dağılım Grafiği (Scatter Plot)

</span>

<span style="color:green">

Dağılım grafiği, bireysel veri noktalarını gösterir ve iki değişken arasındaki ilişkiyi görselleştirmek için kullanılır.

</span>

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(x, np.exp(x));
```

<span style="color:purple">

- **ax.scatter(x, np.exp(x)):** x değerleri ile np.exp(x) (x'in üstel fonksiyonu) değerleri arasında bir dağılım grafiği çizer. np.exp(x) ifadesi, e tabanında üstel fonksiyonunu hesaplar ve her bir x değeri için bu hesaplamayı yapar.

</span>

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(x, np.sin(x));
```

<span style="color:purple">

- **ax.scatter(x, np.sin(x)):** x değerleri ile np.sin(x) (x'in sinüs fonksiyonu) değerleri arasında bir dağılım grafiği çizer. np.sin(x) ifadesi, her bir x değeri için sinüs fonksiyonunu hesaplar.

</span>

<span style="color:blue">

## 3. Sütun Grafiği (Bar Plot)

</span>

<span style="color:green">

Sütun grafiği, kategorik verileri göstermek için kullanılır ve her bir kategori için bir sütun çizer.

- **ax.bar()** ile dikey çubuk grafikleri ve 
- **ax.barh()** ile yatay çubuk grafikleri oluşturabilirsiniz.

</span>

```python
#fiyatları içeren bir dictionary oluşturduk
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

<span style="color:purple">

- **nut_butter_prices:** Farklı fındık ezmesi türlerinin fiyatlarını içeren bir sözlük oluşturur. Sözlüğün anahtarları (keys) ürün adlarını, değerleri (values) ise fiyatları temsil eder.

- **ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()):** Sözlükteki anahtarları (ürün adları) x ekseninde ve değerleri (fiyatlar) y ekseninde kullanarak bir sütun grafiği çizer.

- **ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)"):** Grafiğe bir başlık ve y eksenine bir etiket ekler.


</span>

```python
#alternatif yatay (horizontal) sütun grafiği
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

<span style="color:purple">

- **ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())):** Sözlükteki anahtarları (ürün adları) y ekseninde ve değerleri (fiyatlar) x ekseninde kullanarak bir yatay sütun grafiği çizer.

</span>

<span style="color:blue">

## 4. Histogram Grafiği (Histogram Plot) 

</span>

<span style="color:green">

Histogram grafiği, veri dağılımını görselleştirmek için kullanılır ve verileri belirli aralıklara ayırarak frekanslarını gösterir.

</span>

```python
# Normal Dağılım (Normal Distribution)
x = np.random.randn(1000)  # Normal dağılımdan 1000 rastgele veri çeker
fig, ax = plt.subplots(figsize=(3, 3))

# Histogram grafiği oluştur
ax.hist(x);
```

```python
#Rastgele Dağılım (Random Distribution)

x = np.random.random(1000)  # Rastgele dağılımdan 1000 veri çeker
fig, ax = plt.subplots(figsize=(3, 3))
ax.hist(x);
```

<span style="color:DarkBlue">

### Bir Figürde Birden Fazla Eksen Oluşturma (Subplots Kullanımı)

</span>

<span style="color:green">

Subplots, aynı Figürde birden fazla Eksen (Axes) oluşturmanıza olanak tanır. Bu, bir Figür içinde birden fazla grafiği görselleştirmenizi sağlar. Subplots kullanarak, başlangıçta bir grafikle başladığınız Figürü, gerektiğinde daha fazla grafik ekleyerek ölçeklendirebilirsiniz.

**Subplots Kullanımı**
- **plt.subplots(nrows, ncols):** nrows satır sayısını, ncols sütun sayısını belirler. Bu parametreler çarpılarak toplam Eksen sayısını belirler.


- Örneğin, plt.subplots(nrows=2, ncols=2) ifadesi, 2*2=4 toplam Eksen oluşturur.

</span>

<span style="color:CornflowerBlue ">

- **Seçenek 1: Her Eksen İçin Ayrı Değişken İsimleri Kullanarak Subplots Oluşturma**

</span>


```python
#2 satır ve 2 sütundan oluşan 4 eksen oluşturur ve her eksene ayrı değişken isimleri atar.
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(8, 4))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

<span style="color:CornflowerBlue ">

- **Seçenek 2: Dizin Kullanarak Subplots Oluşturma**

</span>

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))

# Index to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## 1. Pandas DataFrame ile Çizgi Grafiği Oluşturma

</span>

<span style="color:green">

Pandas kütüphanesi, Matplotlib ile entegre olarak doğrudan DataFrame'ler üzerinden grafikler oluşturmayı kolaylaştırır. Bu örnekte, pandas ile zaman serisi verisi oluşturup, kümülatif toplamını alarak çizgi grafiği çizme sürecini inceleyeceğiz.

- **pd.Series(data, index):** Verilen veriyi ve tarih aralığını kullanarak bir pandas Series nesnesi oluşturur.

</span>

<span style="color:CadetBlue">

#### 1. Örnek Zaman Serisi Verisi Oluşturma
</span>

```python
# Örnek zaman serisi verisi oluşturma
# 1/1/2024 tarihinden itibaren 1000 günlük bir tarih aralığı oluşturur.
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2024', periods=1000))

# Zaman serisi verisini göster
print(ts)
```

<span style="color:CadetBlue">

#### 2. Kümülatif Toplamı Hesaplama

- **ts.cumsum():** Zaman serisi verisinin kümülatif toplamını hesaplar. Bu fonksiyon, her adımda önceki değerleri toplayarak yeni bir zaman serisi oluşturur.
</span>

```python
# Değerlerin kümülatif toplamını hesapla
ts_cumsum = ts.cumsum()

# Kümülatif toplamı göster
print(ts_cumsum)
```

<span style="color:CadetBlue">

#### 3. Kümülatif Toplamı Çizgi Grafiği Olarak Görselleştirme
</span>

```python
# Zaman serisi verisini çizgi grafiği olarak görselleştir
ts_cumsum.plot(kind='line')

# Grafiği özelleştir(başlık ekle)ve göster
plt.title('Cumulative Sum Line Plot')
plt.show()
```

<span style="color:MediumVioletRed">
    
### Örnek
</span> 

<span style="color:CadetBlue">

#### 1. Veriyi İçe Aktarma ve Fiyat Kolonunu Temizleme
</span>

```python
car_sales = pd.read_csv("data/car-sales.csv")
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$,\.]', '', regex=True).astype(int)

# DataFrame'i göster
print(car_sales)
```

<span style="color:CadetBlue">

#### 2. Fiyat Kolonundaki Son İki Sıfırı Kaldırma
</span>

```python
#Price kolonundaki her bir değerin son iki karakterini kaldırır.
car_sales["Price"] = car_sales["Price"].astype(str).str[:-2].astype(int)
car_sales
```

<span style="color:CadetBlue">

#### 3. Satış Tarihi Kolonu Ekleme
</span>

```python
#1/1/2024 tarihinden başlayarak DataFrame'in uzunluğu kadar gün sayısını içeren bir tarih aralığı oluşturur ve bunu Sale Date kolonuna ekler.
car_sales["Sale Date"] = pd.date_range("1/1/2024", periods=len(car_sales))
print(car_sales)
```

<span style="color:CadetBlue">

#### 4. Toplam Satışlar Kolonunu Ekleme
</span>

```python
#Price kolonunu integer tipine dönüştürür ve kümülatif toplamını alarak Total Sales kolonuna ekler.
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
print(car_sales)
```

<span style="color:CadetBlue">

####  5. Toplam Satışları Çizgi Grafiği Olarak Görselleştirme

```python
# Yeni bir figür ve eksen(ler) oluştur ve boyutlarını belirle
fig, ax = plt.subplots(figsize=(4, 4))  # Boyutları ayarlayın (örneğin, 10 genişlik ve 6 yükseklik)

# Toplam satışları çizgi grafiği olarak görselleştir
car_sales.plot(x='Sale Date', y='Total Sales', ax=ax)

# Grafiği özelleştir ve göster
ax.set_title('Total Sales Over Time')
plt.show()


##VEYA
#Sale Date kolonunu x ekseni ve Total Sales kolonunu y ekseni olarak kullanarak bir çizgi grafiği çizer.
#car_sales.plot(x='Sale Date', y='Total Sales')

#Grafiğe başlık ekler.
#plt.title('Total Sales Over Time')
#plt.show()
```

## 2. Pandas DataFrame'den Dağılım Grafiği (Scatter) Oluşturma

<span style="color:green">

Pandas kullanarak bir DataFrame'den doğrudan dağılım grafiği (scatter plot) oluşturabiliriz. **kind="scatter"** parametresi ile scatter plot oluşturabiliriz..


</span>

<span style="color:CadetBlue">

#### 1. Fiyat kolonunu string olarak ayarla
</span>

```python
# Not: Matplotlib ve Pandas'ın önceki sürümlerinde, "Price" kolonunun string olarak kalması hata verebilir.
car_sales["Price"] = car_sales["Price"].astype(str)

fig, ax = plt.subplots(figsize=(4, 4))

#Odometer (KM) kolonunu x ekseni ve Price kolonunu y ekseni olarak kullanarak bir dağılım grafiği oluşturur.
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter",ax=ax);

# Grafiği göster
plt.title('Scatter Plot with Price as String')
plt.show()
```

<span style="color:CadetBlue">

#### 2. Fiyat kolonunu integer olarak ayarla
</span>

```python
car_sales["Price"] = car_sales["Price"].astype(int)

fig, ax = plt.subplots(figsize=(4, 4))

# Dağılım grafiği oluştur
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter",ax=ax)

# Grafiği göster
plt.title('Scatter Plot with Price as Integer')
plt.show()
```

<span style="color:purple">

**Neden Önce String Sonra Integer Olarak Ayarladık?**

__2.Veri Temizleme ve Dönüştürme:__

Veriler CSV dosyasından okunurken, sayısal veriler bile bazen string olarak yorumlanabilir. Bu durum, veriler üzerinde matematiksel işlemler veya görselleştirme yaparken sorunlara yol açabilir. Bu yüzden verilerin doğru veri tipine dönüştürülmesi gereklidir.

__3. Farklı Veri Tiplerini Anlamak:__

Farklı veri tiplerinin nasıl davranacağını görmek ve veri işleme adımlarını daha iyi anlamak için Price kolonunu önce string sonra integer yaparak bu dönüşüm süreçlerini göstermek amacıyla yapılmış olabilir.

Pratikte, birçok geliştirici her ikisini de kullanır; hızlı görevler için pyplot ile başlayıp, daha detaylı işler için OO arayüzüne geçiş yaparlar.

<span style="color:MediumVioletRed">
    
### Örnek:50 yaş üstü insanların analizi
Bunu yapmak için, önce pandas'tan doğrudan bir grafik oluşturarak başlayalım, ardından da nesne yönelimli API'yi (plt.subplots()) kullanarak bunu geliştirelim.

</span> 

<span style="color:CadetBlue">

#### Adım 1: 50 Yaş Üzeri Hastaların Filtrelenmesi
İlk olarak, heart_disease DataFrame'inden yaşları 50'nin üzerinde olan hastalar filtrelenilir ve over_50 adlı yeni bir DataFrame oluşturulur, bu dataframe 50 yaş üzerindeki hastaları içerir.
</span>

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

<span style="color:CadetBlue">

#### Adım 2: Pandas ile Scatter Plot Oluşturma
Pandas kullanarak doğrudan bir scatter plot oluşturuluyor. Burada, yaş (age) x ekseninde, kolesterol seviyeleri (chol) y ekseninde ve kalp hastalığı olup olmadığını gösteren target sütunu renklendirme için kullanılıyor.Bu grafik, yaş ve kolesterol seviyeleri arasındaki ilişkiyi gösterir ve target sütunu ile renklendirilmiş olarak kalp hastalığı olup olmadığını belirtir.
</span>

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target", 
             figsize=(10, 4));
```

<span style="color:CadetBlue">

#### Adım 3: Matplotlib ile Grafik Oluşturma
Daha fazla özelleştirme yapabilmek için Matplotlib'in subplots() fonksiyonu kullanılarak bir Figure ve Axes nesnesi oluşturuluyor.

Burada, aynı grafik Matplotlib kullanılarak yeniden oluşturuluyor ve x ekseni sınırları (45, 100) arasında olacak şekilde ayarlanıyor.
</span>

<span style="color:purple">

**x.set_xlim([45, 100])**
- **ax:** Bu, Matplotlib'in Axes nesnesini temsil eder. Grafiğin üzerinde çizileceği eksenleri (axes) kontrol eder.
- **set_xlim:** Bu, x ekseninin limitlerini ayarlayan bir fonksiyondur. Fonksiyon iki parametre alır: minimum ve maksimum x değeri.
- **[45, 100]:** Bu, x ekseni için belirlenen limitlerdir. Bu durumda, x ekseni 45 ile 100 arasında değişecektir.

Bu kodun kullanımı, grafiğin x ekseninde hangi değer aralığının gösterileceğini kontrol etmenize olanak tanır. Örneğin, yaş (age) verilerini gösteren bir grafikte, sadece 45 ile 100 yaş arasındaki verileri görmek istiyorsanız bu limiti ayarlayabilirsiniz.
</span>

```python
fig, ax = plt.subplots(figsize=(10,4))

over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45, 100]);
```

<span style="color:CadetBlue">

#### Adım 4: Gelişmiş Özelleştirmeler ile Grafik Oluşturma
Bu adımda, Axes nesnesi kullanılarak doğrudan scatter plot oluşturuluyor ve renklerle kalp hastalığı durumunu belirtilir.

Grafik, başlık ve eksen etiketleriyle özelleştirilir ve bir lejant eklenir. Bu sayede, grafik görsel olarak daha anlaşılır ve bilgilendirici hale gelir.
</span>

<span style="color:purple">

- **fig:** Grafiğin tamamını kapsayan ana nesnedir.
- **ax:** Grafiğin üzerinde çizileceği eksenleri temsil eder.

**ax.scatter(...):** ax nesnesine bir scatter plot çizer.
* **over_50["age"]:** X ekseninde gösterilecek veri (yaş).
* **over_50["chol"]:** Y ekseninde gösterilecek veri (kolesterol seviyesi).
* **c=over_50["target"]:** Noktaların renklerini belirlemek için kullanılacak veri (target sütunu, kalp hastalığı olup olmadığını gösterir; 0 = kalp hastalığı yok, 1 = kalp hastalığı var).
- **scatter:** Scatter plot nesnesini döndürür, bu da ileride ekleme ve özelleştirme yapmak için kullanılabilir.

**ax.legend(...):** Grafiğe bir lejant (legend) ekler.
- __*scatter.legend_elements():__ Scatter plot için otomatik olarak lejant elemanları oluşturur. __*__ operatörü, fonksiyonun döndürdüğü öğeleri ayrı argümanlar olarak geçirir.

</span>

```python
fig, ax = plt.subplots(figsize=(10, 4))

scatter = ax.scatter(over_50["age"],
                     over_50["chol"],
                     c=over_50["target"])

# Grafiğin başlığını ve eksen etiketlerini ayarlar.
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Setup the Legend
ax.legend(*scatter.legend_elements(), 
          title="Target");
```

<span style="color:CadetBlue">

#### Adım 5: Ortalama Kolesterol Seviyesini Gösteren Çizgi Ekleme
Grafiğe yatay bir çizgi eklenerek ortalama kolesterol seviyesini gösteriyor.
Bu sayede, grafik görsel olarak daha anlaşılır ve bilgilendirici hale gelir, aynı zamanda ortalama kolesterol seviyesinin ne olduğunu da net bir şekilde gösterir.
</span>

<span style="color:purple">

**ax.axhline(...):** Yatay bir çizgi ekler.
-  **over_50["chol"].mean():** chol sütunundaki değerlerin ortalamasını hesaplar ve bu değerde bir yatay çizgi çizer.
- **linestyle="--":** Çizginin stilini kesik çizgi olarak ayarlar.

</span>

```python

fig, ax = plt.subplots(figsize=(10,4 ))

scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), 
          title="Target")
# yatay çizgi ekleme
ax.axhline(over_50["chol"].mean(),
           linestyle="--"); # style the line to make it look nice
```

<span style="color:blue">

### Aynı figürde birden fazla grafiği (plot) göstermek
### Mevcut stillendirilmiş grafiğe başka bir veri seti eklemek.

Kod parçası iki eksen (axes) kullanarak bir figürde iki ayrı scatter plot oluşturur ve her iki grafiğe de ortalama değerleri gösteren yatay çizgiler ekler. Kodun her bir satırının detaylı açıklaması aşağıda verilmiştir:
</span>


<span style="color:CadetBlue">

#### Adım 1: Figür ve Axes Oluşturulması
</span>

<span style="color:CadetBlue">

#### Adım 2: İlk Grafik (ax0): Kalp Hastalığı ve Kolesterol Seviyeleri
</span>

<span style="color:CadetBlue">

#### Adım 3: İkinci Grafik (ax1): Kalp Hastalığı ve Maksimum Kalp Atış Hızı
</span>

<span style="color:purple">

**fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 8)):** Matplotlib'in subplots fonksiyonu kullanılarak bir Figure (fig) ve *iki Axes (ax0 ve ax1)* nesnesi oluşturulur.
- **nrows=2:** Figürde 2 satır olacak şekilde düzenler.
- **ncols=1:** Figürde 1 sütun olacak şekilde düzenler.
- **sharex=True:** Her iki eksenin de aynı x eksenini paylaşmasını sağlar.

</span>

```python
# Setup plot (2 rows, 1 column)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 rows
                               ncols=1, # 1 column 
                               sharex=True, # both plots should use the same x-axis
                               figsize=(10,8))
#----------------------------------------------------------------

# Add data for ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='b',                   # çizgi rengi mavi demek
            linestyle='--', 
            label="Average")  

#----------------------------------------------------------------

# Add data for ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")

# Title the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, 
             fontweight='bold');
```

### c) Renk Haritaları (cmap) ile Grafiklerin Renklerini Özelleştirmek

 <span style="color:green">

**cmap="viridis":** Renk haritası (colormap) olarak "viridis" kullanılır. cmap parametresi, scatter plot'taki noktaların renklerinin hangi skala ile gösterileceğini belirler. "viridis" colormap'i, Matplotlib tarafından sağlanan bir renk skalasıdır ve genellikle veri görselleştirmelerinde güzel renk geçişleri sağladığı için tercih edilir. "viridis" yeşil, mavi ve sarı tonlarından oluşan bir renk paletidir.


**cmap="winter"** parametresini kullanarak grafiğin renk haritasını değiştirebilirsiniz. "winter" colormap'i, Matplotlib tarafından sağlanan bir başka renk skalasıdır ve mavi ve yeşil tonlarından oluşur. Bu, veri noktalarınızı farklı bir renk düzeniyle görselleştirmenize olanak tanır.
